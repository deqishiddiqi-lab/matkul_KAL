#**1. Matriks A**
$$A=\begin{bmatrix}a&b&c&d&e\\f&g&h&i&j\\k&l&m&n&o\\p&q&r&s&t\\u&v&w&x&y\end{bmatrix}$$

#**2.Determinan Matriks 5x5**
Menggunakan ekspansi kofaktor pada baris pertama:
$$det(A)=aC_{11}-bC_{12}+cC_{13}-dC_{13}+eC_{14}$$
dengan:
$$C_{ij}=(-1)^{i+j}M_{ij}$$

#**3.Matriks Adjoin (adj A)**
adj(A) = transpose dari matriks kofaktor
#
Bentuk matriksnya:
$$A=\begin{bmatrix}C_{11}&C_{21}&C_{31}&C_{41}&C_{51}\\C_{12}&C_{22}&C_{32}&C_{42}&C_{52}\\C_{13}&C_{23}&C_{33}&C_{43}&C_{53}\\C_{14}&C_{24}&C_{34}&C_{44}&C_{54}\\C_{15}&C_{25}&C_{35}&C_{45}&C_{55}\end{bmatrix}$$

#**4.invers Matriks**
$$A^{-1}=\frac{1}{det(A)}adj(A)$$

#**5. Syarat invers**
$$det(A)\neq0$$
Jika: det(A) = 0 → A tidak memiliki invers
