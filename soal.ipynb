#A. Hitung determinan matrik berikut dengan menggunakan rumus expansi baris


1.   $$A=\begin{bmatrix}-7&-5\\1&4\end{bmatrix}$$
#jawaban
menggunakan ekspansi baris pertama(i=1):
#
$a_{11}=-7,M_{11}=4$
#
$a_{12}=-5,M_{12}=1$
#
$det(A)=(-1)^{1+1}(-7)(4)+(-1)^{1+2}(-5)(1)$
#
$det(A)=(1)(-28)+(-1)+(-5)$
#
$det(A)=-28+-5=-23$

2.   $$A=\begin{bmatrix}0&2&-3\\1&-2&-1\\0&0&1\end{bmatrix}$$
#jawaban
karena angka nol terbanyak ada di baris ke tiga maka saya akan mengambil baris 3(i=3)
#
$a_{31}=0,a_{32}=0$
#
$a_{33}=1,M_{32}=A=\left|\begin{bmatrix}0&2\\1&-2&\end{bmatrix}\right| =(0)(-2)-(2)(1)=-2$
#
$det(A)=(-1)^{3+3}(1)(-2)=(1)(1)(-2)=-2$
#

#B. gunakan rumus matriks adjoin untuk menghitung invers dari matriks berikut dengan rumus
1.   $$A=\begin{bmatrix}-7&-5\\1&4\end{bmatrix}$$
#jawaban
det(A)=-23 ini hasil dari soal nomor 1
#
$C_{11}=4,C_{12}=-1,C_{21}=5,C_{22}=-7$
#
$adj(A)=\begin{bmatrix}4&5\\-1&-7\end{bmatrix}$
#
$A^{-1}=\frac{1}{-23}\begin{bmatrix}4&5\\-1&-7\end{bmatrix}=\begin{bmatrix}-4/23&-5/23\\1/23&7/23\end{bmatrix}$

2.   $$A=\begin{bmatrix}0&2&-3\\1&-2&-1\\0&0&1\end{bmatrix}$$
#jawaban
$C_{11}=+|(-2)(1)-(0)(-1)|=-2$
#
$C_{12}=-|(1)(1)-(0)(-1)|=-1$
#
$C_{13}=+|(1)(0)-(0)(-2)|=0$
#
$C_{21}=-|(2)(1)-(0)(-3)|=-2$
#
$C_{22}=+|(0)(1)-(0)(-3)|=0$
#
$C_{23}=-|(0)(0)-(0)(2)|=0$
#
$C_{31}=+|(2)(-1)-(-2)(-3)|=-8$
#
$C_{32}=-|(0)(-1)-(-1)(-3)|=-3$
#
$C_{33}=+|(0)(-2)-(1)(2)|=-2$
#
$adj(A)=C^T=\begin{bmatrix}-2&-2&-8\\-1&0&-3\\0&0&-2\end{bmatrix}$
#
$=A^{-1}=\frac{1}{-2}\begin{bmatrix}-2&-2&-8\\-1&0&-3\\0&0&-2\end{bmatrix}=\begin{bmatrix}1&1&4\\0,5&0&1,5\\0&0&1\end{bmatrix}$
