```python
bentuk_matriks=[["1 1 1 1 1 |5"],["2 3 1 1 1 |8"],["1 1 2 1 1 |7"],["1 1 1 3 1 |9"],["1 1 1 1 2 |8"]]
for n in bentuk_matriks:
  print(n)
```

#langkah 1
hilangkan elemen pivot pertama


*   R<sub>2</sub>= R<sub>2</sub>-2R<sub>1</sub>
*   R<sub>3</sub>= R<sub>3</sub>-R<sub>1</sub>
*   R<sub>4</sub>= R<sub>4</sub>-R<sub>1</sub>
*   R<sub>5</sub>= R<sub>5</sub>-R<sub>1</sub>




```python
langkah_1 =[["1 1 1 1 1 |5"],["0 1 -1 -1 -1 |-2"],["0 0 1 0 0 |2"],["0 0 0 2 0 |4"],["0 0 0 0 1 |3"]]
for m in langkah_1:
  print(m)
```
